# Demo

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
type(iris)
```

```python
import pandas as pd
data = pd.DataFrame({
    'sepal length': iris.data[:, 0],
    'sepal width': iris.data[:, 1],
    'petal length': iris.data[:, 2],
    'petal width': iris.data[:, 3],
    'species': iris.target
})
data.head()
```

```python
X = data[['petal length', 'petal width']]
y = data['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Cross Validation - Xác thực chéo

### Phương pháp train test split

```python
# Thuc hien nhieu lan cho cong viec
# them 1 vong lap chay 10 lan cho doan code day:
# dung dictrionary de luu tru ket qua
# for i in range (1, 11)
# 70%, 75%, 80% training
test_size_list = [0.3, 0.25, 0.2]
for test_size in test_size_list:
  X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y, test_size=test_size)
  clf1 = RandomForestClassifier(n_estimators=100)
  clf1.fit(X_train_1, y_train_1)

  score_train = clf1.score(X_train_1, y_train_1)
  score_test = clf1.score(X_test_1, y_test_1)

  print("With [", 1 - test_size, ":", test_size, "], scorer train is", round(score_train, 2),
        ", score test is", round(score_test, 2),
        "diff is", round(abs(score_train - score_test), 2))
  
# compare between results
# choose the best one
```

### Phương pháp K-folds

```python
X.head()
```

```python
y.values
```

```python
from sklearn.model_selection import KFold
from sklearn import model_selection
```

```python
clf_k = RandomForestClassifier(n_estimators=100)
kfold = KFold(n_splits=15)
results = model_selection.cross_val_score(clf_k, X, y, cv = kfold)
print("Accuracy %.2f%% (%.2f%%)" % (results.mean() * 100, results.std()* 100))
```

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
param_grid = {
    'n_estimators': [30, 50, 100, 150, 200],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}
```

```python
from datetime import datetime
from datetime import timedelta
```

```python
start_time = datetime.now()
```

```python
CV_rfc = GridSearchCV(estimator=RandomForestClassifier(),
                      param_grid=param_grid, cv = 10)
```

```python
CV_rfc.fit(X_train_1, y_train_1)
```

```python
end_time = datetime.now()
result = end_time - start_time
```

```python
result
```

```python
print(CV_rfc.best_params_)
```

```python
y_pred_1 = CV_rfc.predict(X_test_1)
```

```python
from sklearn import metrics
```

```python
print("Accuracy", metrics.accuracy_score(y_test_1, y_pred_1))
```

```python
CV_rfc.predict([[5, 2]])
```
